```python
import pandas as pd

df= pd.read_csv ('./datasets/attacks.csv', encoding='unicode_escape')
df
```

```python
## isolate fatal attacks and dropna 

df.columns
FSsub= df[['Fatal (Y/N)', 'Species ', 'Country']].dropna()
fatalattacks= FSsub[FSsub['Fatal (Y/N)']== 'Y']
fatalattacks
list(fatalattacks['Species '])
```

```python
## Get list of species 

fatalattacks['Species ']
fatalattacks['fatals']= fatalattacks['Species '].str.extract(r'(\w{3,}\sshark)|(?:\w{3,}\spointer)')
```

```python
fatalattacks['Species '].replace('Blue pointer', 'blue pointer shark', inplace= True)
fatalattacks['Species '].replace('Blue pointers', 'blue pointer shark', inplace= True)
```

```python
fatalattacks[fatalattacks['Species ']== 'whaler shark']
```

```python
## Create correct dataframe

fatalattacks['Fatal_Species']= fatalattacks['Species '].str.extract(r'(\w{3,}?-\w{3,}\sshark|\w{3,}?\s\w{3,}\sshark|\w{3,}\sshark)')
fataldf= fatalattacks[['Fatal (Y/N)', 'Fatal_Species', 'Country']]


```

```python
fataldf= fataldf.dropna()

fataldf.applymap(lambda x: x.title())

```
